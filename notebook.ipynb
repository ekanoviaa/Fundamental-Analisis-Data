## Menentukan Pertanyaan Bisnis

**SMART Question** adalah sebuah framework untuk merumuskan pertanyaan secara terstruktur agar memeperoleh informasi yang mendalam.

* **Spesific (Spesifik)**
  * Pertanyaan harus jelas, fokus pada sebuah topik tertentu, dan tidak bermakna ganda. Hindari pertanyaan yang terlalu luas.
    * Salah: Bagaimana cara meningkatkan penjualan?
    * Benar: Faktor apa saja yang memengaruhi penurunan penjualan produk kategori elektronik di wilayah Jakarta selama kuartal terakhir?
* **Measurable (Terukur)**
  * Pertanyaan harus bisa dijawab dengan angka atau matrix yang konkret, harus tahu apa yang akan dihitung.
    * Salah: Apakah pelanggan senang dengan layanan kita?
    * Benar: Berapa skor rata-rata Customer Satisfaction untuk layanan purna jual bulan ini dibandingkan bulan lalu?
* **Action-Oriented (Berorientasi Aksi)**
  * Hasil dari pertanyaan harus bisa memberikan arahan untuk melakukan tindakan nyata. Jika pertanyaan terjawab, stakeholder harus tahu apa langkah selanjutnya.
    * Salah: Mengapa orang suka berbelanja?
    * Benar: Fitur apa pada aplikasi yang paling sering digunakan sebelum pengguna memutuskan untuk melakukan checkout?
* **Relevant (Relevan)**
  * Hasil dari pertanyaan harus sejalan dengan tujuan utama bisnis atau masalah yang sedang dihadapi.
    * Salah: Menanyakan tentang stok gudang saat masalah utamanya adalah efektivitas kampanye media sosial.
    * Benar: Apakah kampanye iklan di Instagram memberikan Return on Ad Spend (ROAS) yang lebih tinggi dibandingkan iklan di TikTok?
* **Time-bound (Terikat Waktu)**
  * Pertanyaan harus ada batasan waktu yang jelas agar analisis memiliki konteks yang tepat.
    * Salah: Berapa banyak pengguna baru kita?
    * Benar: Berapa tingkat pertumbuhan pengguna baru secara bulanan (Month-over-Month) sepanjang tahun 2025?

## Data Wrangling

### Gathering Data

#### Load df Bike Sharing Dataset

```python
day_df = pd.read_csv('day.csv')
hour_df = pd.read_csv('hour.csv')
```

**Insight:** (Opsional)

- Dataset hour.csv memiliki 17.379 baris yang mencatat data per jam. Dataset ini adalah kunci untuk memahami perilaku pengguna pada jam-jam spesifik (seperti jam sibuk pagi vs sore).
- Kedua dataset memiliki fitur yang kaya, mulai dari kondisi cuaca (suhu, kelembapan) hingga tipe hari (hari libur, hari kerja).

### Assessing Data

#### Identifying Bike Sharing Dataset problem

```python
# Cek struktur data dan tipe data
print(day_df.info())

# Cek apakah ada nilai yang kosong (missing values)
print("\nMissing values pada day_df:\n", day_df.isna().sum())

# Cek apakah ada data duplikat
print("\nJumlah duplikasi pada day_df:", day_df.duplicated().sum())

# Cek statistik deskriptif
print("\nDeskripsi Statistik:\n", day_df.describe())
```

**Steps to Take:**
-  Kesalahan Tipe Data: Kolom dteday harus diubah dari object menjadi datetime.
- Keterbacaan Kategorikal: Nilai angka pada season, yr, dan weathersit perlu diubah ke label deskriptif agar visualisasi mudah dipahami.
- Outlier/Inaccurate Value: Terdapat nilai hum (kelembapan) yang bernilai 0 pada day_df, yang tidak mungkin secara alami di Washington D.C.

**Insight:** (Opsional)
- Data bersih dari missing values (0 null), yang menandakan kualitas perekaman data oleh sistem Capital Bikeshare sangat baik.
- Tidak ditemukan data duplikat, sehingga tidak perlu dilakukan pembersihan baris ganda.
- Beberapa variabel seperti temp dan hum sudah dalam bentuk ternormalisasi (0 hingga 1), sehingga untuk analisis interpretatif, kita perlu memahami skala aslinya (misal: suhu maksimal 41°C).

### Cleaning Data

#### Fixing Bike Sharing Dataset problem

```python
# Load data
day_df = pd.read_csv('day.csv')
hour_df = pd.read_csv('hour.csv')

# Mengubah tipe data dteday ke datetime
day_df['dteday'] = pd.to_datetime(day_df['dteday'])
hour_df['dteday'] = pd.to_datetime(hour_df['dteday'])

# Handling Inaccurate Value (Kelembapan 0 diubah ke rata-rata)
day_df['hum'] = day_df['hum'].replace(0, day_df['hum'].mean())

# Mapping label kategorikal agar lebih informatif
day_df['season'] = day_df['season'].map({1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'})
day_df['yr'] = day_df['yr'].map({0: '2011', 1: '2012'})
day_df['weathersit'] = day_df['weathersit'].map({
    1: 'Clear', 2: 'Misty', 3: 'Light Rain/Snow', 4: 'Heavy Rain/Snow'
})

# Menampilkan hasil cleaning
day_df.head()
```

**Insight:** (Opsional)
- Dataset lebih mudah dipahami
- Normalisasi tipe data tanggal memungkinkan kita melakukan filter berdasarkan rentang waktu spesifik (seperti hanya tahun 2012) tanpa kesulitan teknis.
Penanganan nilai hum yang 0 dengan metode mean imputation menjaga distribusi data tetap stabil tanpa menghilangkan informasi penting dari baris tersebut

## Exploratory Data Analysis (EDA)

### Explore ...

```python
# Mengeksplorasi rata-rata penyewaan berdasarkan jam pada hari kerja vs akhir pekan
# Menggunakan hour_df karena membutuhkan detail jam (hr)
hourly_workingday = hour_df.groupby(['workingday', 'hr'])['cnt'].mean().reset_index()

# Mengeksplorasi dampak suhu ekstrem terhadap tipe pengguna di musim panas
# Filter musim panas (Summer) dan suhu denormalisasi (temp * 41)
summer_df = day_df[day_df['season'] == 'Summer'].copy()
summer_df['temp_celcius'] = summer_df['temp'] * 41

# Korelasi sederhana
print("Korelasi Suhu dengan Casual:", summer_df['temp_celcius'].corr(summer_df['casual']))
print("Korelasi Suhu dengan Registered:", summer_df['temp_celcius'].corr(summer_df['registered']))
```

**Insight:** (Opsional)
- Pengguna registered menunjukkan ketergantungan pada jam tertentu (komuter).
- Pengguna casual memiliki sensitivitas yang lebih tinggi terhadap perubahan suhu dibandingkan pengguna terdaftar.

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
# Data tahun 2012
hour_2012 = hour_df[hour_df['yr'] == 1]
hourly_trend = hour_2012.groupby(['workingday', 'hr'])['cnt'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=hourly_trend, x='hr', y='cnt', hue='workingday', marker='o')
plt.title('Pola Penyewaan Sepeda Per Jam: Hari Kerja vs Akhir Pekan (2012)')
plt.xticks(range(0, 24))
plt.xlabel('Jam')
plt.ylabel('Rata-rata Penyewaan')
plt.legend(title='Keterangan', labels=['Akhir Pekan/Libur', 'Hari Kerja'])
plt.show()
```

### Pertanyaan 2:

```python
summer_df = day_df[day_df['season'] == 'Summer'].copy()
summer_df['temp_celcius'] = summer_df['temp'] * 41

# Filter hanya hari-hari dengan suhu ekstrem (>30°C)
hot_summer_days = summer_df[summer_df['temp_celcius'] > 30]

# Hitung rata-rata penyewaan untuk masing-masing tipe pengguna
mean_rentals = hot_summer_days[['casual', 'registered']].mean().reset_index()
mean_rentals.columns = ['User_Type', 'Avg_Rentals']

# Membuat Visualisasi Diagram Batang
plt.figure(figsize=(10, 6))
sns.barplot(data=mean_rentals, x='User_Type', y='Avg_Rentals', palette='Oranges')

# Menambahkan judul dan label sumbu yang jelas
plt.title('Rata-rata Penyewaan Sepeda saat Suhu Ekstrem (>30°C) di Musim Panas')
plt.xlabel('Tipe Pengguna')
plt.ylabel('Rata-rata Jumlah Penyewaan (Harian)')

# Menampilkan grafik
plt.show()
```

**Insight:** (Opsional)
- Visualisasi diagram batang ini memberikan perbandingan instan dan mudah dibaca. Terlihat jelas bahwa batang untuk pengguna Registered jauh lebih tinggi (lebih dari 3x lipat) dibandingkan batang untuk pengguna Casual ketika suhu udara sangat panas (>30°C).
- Rata-rata penyewaan harian pengguna terdaftar berada di angka yang signifikan (sekitar 3500-an), sementara pengguna kasual hanya berada di angka rendah (di bawah 1000). Ini mengonfirmasi tren utama bahwa suhu panas ekstrem tidak menghentikan pengguna terdaftar untuk bersepeda.
- Kesimpulan bahwa kedua kelompok pengguna memiliki motivasi yang berbeda. Pengguna terdaftar kemungkinan besar adalah komuter yang "terpaksa" bersepeda untuk bekerja, sementara pengguna kasual yang bersepeda untuk rekreasi lebih memilih untuk membatalkan niat mereka saat cuaca tidak nyaman.

## Analisis Lanjutan (Opsional)
